## üõ†Ô∏è Methodology / Pipeline

### 1Ô∏è‚É£ Data Collection & Preparation
- Collect survey responses, support tickets, or product feedback  
- Clean and preprocess text: remove extra spaces, normalize punctuation, correct minor errors  
- Combine relevant fields (e.g., title + description) for embedding context  

**Benefit:** Ensures data quality and consistent AI performance  

### 2Ô∏è‚É£ Text Embeddings
- Use Hugging Face **SentenceTransformer models** (`all-MiniLM-L6-v2`) to convert text into vectors  
- Captures semantic meaning so similar ideas are close in vector space  

**Benefit:** Understands meaning beyond simple keywords  

### 3Ô∏è‚É£ Vector Search with FAISS
- Build a **FAISS index** from embeddings for fast similarity search  
- Query new input ‚Üí retrieve top-k similar items  

**Benefit:** Rapid retrieval of semantically similar entries  

### 4Ô∏è‚É£ Insight Generation
- Hugging Face **summarization models** (`sshleifer/distilbart-cnn-12-6`) generate concise recommendations from top matches  

**Benefit:** Produces actionable, human-readable insights

### ü§ñ Automated Ticket Triage with Sentence-BERT & FAISS

- **Purpose:** Automatically recommend resolutions for incoming support tickets based on historical tickets.  
- **Workflow:**  
  1. Combine each ticket's `title` and `description` into a single text field.  
  2. Generate embeddings using `sentence-transformers/all-MiniLM-L6-v2`.  
  3. Normalize embeddings and build a **FAISS index** for fast similarity search.  
  4. For a new incoming ticket, retrieve the top-K most similar historical tickets.  
  5. Recommend a resolution using majority vote from top matches.  
- **Outputs:**  
  - List of top matching tickets with similarity scores.  
  - Recommended resolution for the incoming ticket.  

This approach helps support teams quickly identify solutions for common issues using AI-powered similarity matching.
  


````python
# ------------------------------
# Install required packages (run once)
# ------------------------------
# In Colab or a fresh env uncomment and run:
# !pip install -q sentence-transformers faiss-cpu pandas

# ------------------------------
# Imports
# ------------------------------
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
import faiss
from collections import Counter

# ------------------------------
# 1) Example historical ticket dataset
# Replace this with your real CSV with columns: ticket_id, title, description, resolution
# ------------------------------
data = [
    {"ticket_id": "T001", "title": "Cannot login", "description": "User reports login failing with 'invalid credentials' message.", "resolution": "Reset password and clear cache; guided user to login"},
    {"ticket_id": "T002", "title": "App crashes on launch", "description": "App crashes immediately after opening on Android 12.", "resolution": "Collect logs and escalate to mobile team; provided temp APK"},
    {"ticket_id": "T003", "title": "Password reset not working", "description": "Forgot password flow returns error when submitting email.", "resolution": "Fixed backend email sending service and retried reset"},
    {"ticket_id": "T004", "title": "Cannot login - MFA", "description": "User cannot complete MFA step, SMS code not received.", "resolution": "Enabled backup codes and advised carrier check"},
    {"ticket_id": "T005", "title": "Slow performance dashboard", "description": "Dashboard queries time out for large reports between 2-4 PM.", "resolution": "Optimized query indexes; scheduled batch processing off-peak"},
    {"ticket_id": "T006", "title": "App crashes on launch - specific device", "description": "Crashes on launch for Samsung S21 after latest OS update.", "resolution": "Reproduced crash, rolled back offending library; pushed hotfix"}
]
df = pd.DataFrame(data)

# If you have a CSV, load it instead:
# df = pd.read_csv("/path/to/historical_tickets.csv")

# ------------------------------
# 2) Prepare text to embed
# We create a single text field per ticket (title + description + (optionally) resolution)
# ------------------------------
def ticket_to_text(row):
    return (row["title"] + ". " + row["description"]).strip()

df["text"] = df.apply(ticket_to_text, axis=1)

# ------------------------------
# 3) Create embeddings (SentenceTransformer)
# Use a small, fast model for production demos: all-MiniLM-L6-v2
# ------------------------------
model_name = "sentence-transformers/all-MiniLM-L6-v2"
model = SentenceTransformer(model_name)

# embed all historical tickets
embeddings = model.encode(df["text"].tolist(), show_progress_bar=True, convert_to_numpy=True)

# normalize embeddings (optional, helps cosine sim via inner product)
def normalize_embeddings(emb):
    norms = np.linalg.norm(emb, axis=1, keepdims=True)
    return emb / np.clip(norms, a_min=1e-9, a_max=None)

embeddings_norm = normalize_embeddings(embeddings).astype("float32")

# ------------------------------
# 4) Build FAISS index for fast nearest-neighbor search
# Using inner-product on normalized vectors implements cosine similarity
# ------------------------------
d = embeddings_norm.shape[1]
index = faiss.IndexFlatIP(d)    # inner product index
index.add(embeddings_norm)      # add vectors

# keep mapping from index position -> ticket info (we'll use df)
# (df row order corresponds to embeddings order)
print(f"Indexed {index.ntotal} tickets (embedding dim={d})")

# ------------------------------
# 5) Query: new incoming ticket
# ------------------------------
def triage_ticket(query_text, top_k=3):
    # 1) embed the query
    q_emb = model.encode([query_text], convert_to_numpy=True)
    q_emb = normalize_embeddings(q_emb).astype("float32")

    # 2) search FAISS
    distances, indices = index.search(q_emb, top_k)  # distances = inner product scores (cosine)
    scores = distances[0].tolist()
    indices = indices[0].tolist()

    # 3) fetch matched tickets and resolutions
    matches = []
    for idx, score in zip(indices, scores):
        if idx < 0:
            continue
        row = df.iloc[idx]
        matches.append({
            "ticket_id": row["ticket_id"],
            "title": row["title"],
            "description": row["description"],
            "resolution": row["resolution"],
            "score": float(score)
        })

    # 4) simple recommendation: majority vote on resolutions among top matches
    resolutions = [m["resolution"] for m in matches if m.get("resolution")]
    if len(resolutions) == 0:
        recommended = None
    else:
        recommended = Counter(resolutions).most_common(1)[0][0]

    return {"query": query_text, "matches": matches, "recommended_resolution": recommended}

# ------------------------------
# Example: incoming ticket
# ------------------------------
incoming = "User cannot sign in ‚Äî gets 'invalid credentials' although password is correct. Also uses MFA sometimes."
result = triage_ticket(incoming, top_k=4)

# pretty print result
from pprint import pprint
print("\n=== Incoming Ticket ===")
print(incoming)
print("\n=== Top Matches ===")
pprint(result["matches"])
print("\n=== Recommended Resolution ===")
pprint(result["recommended_resolution"])
````
